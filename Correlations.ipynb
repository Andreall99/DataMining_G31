```python
heatmap_list = columns_dict["numerical_columns"]
heatmap_list = heatmap_list.drop(['DOW_0', 'DOW_1', 'DOW_2',
       'DOW_3', 'DOW_4', 'DOW_5', 'DOW_6', 'HR_0', 'HR_1', 'HR_2', 'HR_3',
       'HR_4', 'HR_5', 'HR_6', 'HR_7', 'HR_8', 'HR_9', 'HR_10', 'HR_11',
       'HR_12', 'HR_13', 'HR_14', 'HR_15', 'HR_16', 'HR_17', 'HR_18', 'HR_19',
       'HR_20', 'HR_21', 'HR_22', 'HR_23'])

heatmap_list
```

```python
import seaborn as sns

plt.figure(figsize=(20, 8))
df[heatmap_list].corr()
import matplotlib.pyplot as plt
# Generate a heatmap
corr_matrix = df[heatmap_list].corr().round(1)
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

There are 3 paired of variables that are highly correlated:

Vendor Count <====> Is Chain

Vendor Count <====> Product Count

Product Count <====> Is Chain

```python
grouped_df = df.groupby(['vendor_count']).agg({'is_chain':'mean', 'product_count': 'mean'}).reset_index()
grouped_df
```

```python
# Create the figure
plt.figure(figsize=(10, 6))

# Create a scatter plot (bubble plot) with averages of Total Money Spent
sns.scatterplot(
    x='vendor_count',  # X-axis - Recency
    y='is_chain',  # Y-axis - Average Order Duration
    size='product_count',  # Bubble size based on the average money spent
    hue='product_count',  # Optional: Color based on average money spent
    data=grouped_df, 
    sizes=(20, 200),  # Control bubble size range
    palette='coolwarm',  # Color palette
    alpha=0.6,  # Transparency
    edgecolor="black"  # Border color of bubbles
)

# Adding labels and title
plt.xlabel('Number of Vendors')
plt.ylabel('Total of times the bought from a chain restaurant')
plt.title('Total Product Count')

# Show the plot
plt.show()
```

Customers tend to rely more on chain restaurants when ordering from multiple vendors, possibly due to convenience, familiarity, and brand recognition.
