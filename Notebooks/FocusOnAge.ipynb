## 2 distribution and comparation with other features

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

sns.violinplot(data=df, x='customer_age', ax=axes[0, 0],color='seagreen')
percentile_99 = df['customer_age'].quantile(0.99)
axes[0, 0].axvline(percentile_99, color='red', linestyle='dotted')
axes[0, 0].set_title('Customer Age')

age_vendor_mean = df.groupby('customer_age')['vendor_count'].mean().reset_index()
sns.scatterplot(data=age_vendor_mean, x='customer_age', y='vendor_count', ax=axes[0, 1],color='seagreen')
axes[0, 1].axvline(percentile_99, color='red', linestyle='dotted')
axes[0, 1].set_title('Vendor Count by Customer Age')

age_order_value_mean = df.groupby('customer_age')['avg_order_value'].mean().reset_index()
sns.scatterplot(data=age_order_value_mean, x='customer_age', y='avg_order_value', ax=axes[1, 0],color='seagreen')
axes[1, 0].axvline(percentile_99, color='red', linestyle='dotted')
axes[1, 0].set_title('Average Order Value by Customer Age')

age_value_freq = df.groupby('customer_age')['value_freq'].mean().reset_index()
sns.scatterplot(data=age_value_freq, x='customer_age', y='value_freq', ax=axes[1, 1],color='seagreen')
axes[1, 1].axvline(percentile_99, color='red', linestyle='dotted')
axes[1, 1].set_title('Value Frequency by Customer Age')

plt.tight_layout()
plt.show()
```

The customer age distribution leans heavily towards the younger demographic. This can lead to potential misinterpretations when analyzing the visual representation of our data. It's crucial to remember that 99% of our customer base is under 52, which explains the more scattered results for the 50 to 80 segment of our graph.

The insights we can glean from these charts are that nearly 99% of our user base is younger than 50, but age does not appear to significantly influence spending frequency or habit

## 3 confidence interval plot for the customer_age

```python

from scipy import stats

age_order_value_mean = df.groupby('customer_age')['avg_order_value'].mean().reset_index()
age_order_value_std = df.groupby('customer_age')['avg_order_value'].std().reset_index()
age_order_value_count = df.groupby('customer_age')['avg_order_value'].count().reset_index()

age_order_value_mean['std'] = age_order_value_std['avg_order_value']
age_order_value_mean['count'] = age_order_value_count['avg_order_value']

age_order_value_mean['standard_error'] = age_order_value_mean['std'] / np.sqrt(age_order_value_mean['count'])

age_order_value_mean['margin_of_error'] = age_order_value_mean['standard_error'] * stats.t.ppf(1-0.05/2, age_order_value_mean['count'] - 1)

age_order_value_mean['confidence_interval'] = age_order_value_mean['margin_of_error'] * 2

fig, ax = plt.subplots(figsize=(15, 5))
sns.scatterplot(data=age_order_value_mean, x='customer_age', y='avg_order_value', ax=ax)
ax.errorbar(age_order_value_mean['customer_age'], age_order_value_mean['avg_order_value'], yerr=age_order_value_mean['confidence_interval'], fmt='o', color='black', ecolor='seagreen', elinewidth=3, capsize=0)

plt.show()
```

```python
#make a barplot of the average age by region, wirh error bars
age_region_mean = df.groupby('customer_region')['customer_age'].mean().reset_index()
age_region_std = df.groupby('customer_region')['customer_age'].std().reset_index()
age_region_count = df.groupby('customer_region')['customer_age'].count().reset_index()

age_region_mean['std'] = age_region_std['customer_age']
age_region_mean['count'] = age_region_count['customer_age']

age_region_mean['standard_error'] = age_region_mean['std'] / np.sqrt(age_region_mean['count'])

age_region_mean['margin_of_error'] = age_region_mean['standard_error'] * stats.t.ppf(1-0.05/2, age_region_mean['count'] - 1)

age_region_mean['confidence_interval'] = age_region_mean['margin_of_error'] * 2

fig, ax = plt.subplots(figsize=(15, 5))
sns.barplot(data=age_region_mean, x='customer_region', y='customer_age', ax=ax, color= 'seagreen')
ax.errorbar(age_region_mean.index, age_region_mean['customer_age'], yerr=age_region_mean['confidence_interval'], fmt='-', color='black', ecolor='red', elinewidth=3, capsize=0)

plt.title("Average age by Region")

plt.show()
```
